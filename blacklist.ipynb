```python
def get_recipes_raw(ingridient):
    file = open(r"C:\\Users\\divit\\cookbook\\recipes\\ingridient_alf.txt".replace("_alf", str(ingridient)), "r")
    array = [row.strip() for row in file]
    if array[2:] != []:
        file.close()
        return array[2:]
    else:
        print('Denchik, tvoyu divisiyu!')
    file.close()

def get_recipes(ingridient):
    array = get_recipes_raw(ingridient)
    array1 = []
    for i in array:
        i = int(i)
        array1.append((df.iloc[i][2]))
    return(array1)

def if_ingridient_in_recipe_raw(ingridient, recipe):
    if recipe in get_recipes_raw(ingridient):
        return 1
    else:
        return 0
    
def get_recipes_many(ingridients):
    array = []
    for i in ingridients.split():
        array += get_recipes(df.columns.get_loc(i))
    array = set(array)
    for i in ingridients.split():
        array = array & set(get_recipes(df.columns.get_loc(i)))
    return array
```
